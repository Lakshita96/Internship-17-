##### Q10: Go to webpage https://www.amazon.in/ Enter “Laptop” in the search field and then click the search icon. Then set CPU Type filter to “Intel Core i7” and “Intel Core i9”. After setting the filters scrape first 10 laptops data. You have to scrape 3 attributes for each laptop:
1. title
2. Ratings
3. Price

```python
driver.get('https://www.amazon.in')
```

```python
search = driver.find_element_by_id('twotabsearchtextbox')
search.send_keys("Laptop")
btn = driver.find_element_by_id('nav-search-submit-button')
btn.click()
```

```python
intelcorei7 = driver.find_element_by_xpath('//*[@id="p_n_feature_thirteen_browse-bin/12598163031"]/span/a/div')
intelcorei7.click()
```

```python
intelcorei9 = driver.find_element_by_xpath('//*[@id="p_n_feature_thirteen_browse-bin/16757432031"]/span/a')
intelcorei9.click()
```

```python
titles = driver.find_elements_by_xpath("//h2[@class='a-size-mini a-spacing-none a-color-base s-line-clamp-2']")
laptop = []
for i in titles:
    laptop.append(i.text)
laptop
```

```python
prc = driver.find_elements_by_xpath("//span[@class='a-price-whole']")
price = []
for i in prc:
    price.append(i.text)
price
```

```python
Laptops = pd.DataFrame({})
Laptops['Brand Name'] = laptop[:10]
Laptops['Price on Amazon'] = price[:10]
Laptops
```
