### 1. Scrape the details of most viewed videos on YouTube from Wikipedia:
- Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos/
You need to find following details:
- A) Rank
- B) Name
- C) Artist
- D) Upload date
- E) Views__

```python
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
```

```python
rank = []

r = driver.find_elements_by_xpath('//*[@id="mw-content-text"]/div[1]/table[3]/tbody/tr/td')

for i in r:
    try:
        rank.append(i.text)
    except NoSuchElementException:
        rank.append('-')
        
#rank
```

```python
Names = []
for i in range(1,len(rank),6):
       Names.append(rank[i])
#Names
```

```python
Rank = []
for i in range(0,len(rank),6):
       Rank.append(rank[i])
#Rank
```

```python
Artist = []
for i in range(2,len(rank),6):
       Artist.append(rank[i])
#Artist
```

```python
Date = []
for i in range(4,len(rank),6):
    Date.append(rank[i])
#Date
```

```python
Views = []
for i in range(3,len(rank),6):
    Views.append(rank[i])
#Views
```

```python
most_viewed_videos_on_youtube = pd.DataFrame({})
most_viewed_videos_on_youtube['Rank'] = Rank
most_viewed_videos_on_youtube['Title of the Video'] = Names
most_viewed_videos_on_youtube['Artist of the Video'] = Artist
most_viewed_videos_on_youtube['Date of Upload'] = Date
most_viewed_videos_on_youtube['Number of views (In Billions)'] = Views
most_viewed_videos_on_youtube
```

### 2. Scrape the details team India’s international fixtures from bcci.tv.

- D) Date
- E) Time
- Note: - From bcci.tv home page you have reach to the international fixture page through code.

```python
driver.get("https://www.bcci.tv/")
```

```python
link = driver.find_element_by_xpath("//a[@class='navigation__link navigation__link--in-drop-down']").get_attribute('href')
```

```python
driver.get(link)
```

```python
Series = []
sr = driver.find_elements_by_xpath("//span[@class='u-unskewed-text fixture__tournament-label u-truncated']")
for i in sr:
    Series.append(i.text)
#Series
```

```python
Title = []
tt = driver.find_elements_by_xpath("//strong[@class='fixture__name fixture__name--with-margin']")
for i in tt:
    Title.append(i.text)
#Title
```

```python
Place = []
loc = driver.find_elements_by_xpath("//p[@class='fixture__additional-info']//span")
for i in loc:
    Place.append(i.text)
#Place
```

```python
Date = []
dt = driver.find_elements_by_xpath("//div[@class='fixture__datetime desktop-only']")
for i in dt:
    Date.append(i.text.split('\n'))
#Date
```

```python
Time = []
tm = driver.find_elements_by_xpath("//span[@class='fixture__time']")
for i in tm:
    Time.append(i.text)
#Time
```

```python
dt = ((Date[0][1] +'-' +Date [0][2]),
      (Date[1][1] +'-' +Date [1][2]),
      (Date[2][1] +'-' +Date [2][2]),
      (Date[3][1] +'-' +Date [3][2]),
      (Date[4][1] +'-' +Date [4][2]))
#dt
```

```python
Day =(Date[0][0],
     Date[1][0],
     Date[2][0],
     Date[3][0],
     Date[4][0]) 
#Day
```

```python
Internation_Fixtures = pd.DataFrame({})
Internation_Fixtures['Match Title'] = Title
Internation_Fixtures['Series'] = Series
Internation_Fixtures['Location'] = Place
Internation_Fixtures['Date'] = dt
Internation_Fixtures['Day'] = Day
Internation_Fixtures['Time'] = Time
Internation_Fixtures
```
